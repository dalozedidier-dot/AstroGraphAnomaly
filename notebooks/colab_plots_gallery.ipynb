## Plots gallery + export (Colab)

Ce notebook:

1. Monte Google Drive (optionnel)
2. Affiche tous les PNG d'un dossier (Drive ou repo)
3. Zipper un dossier et le télécharger


````python
from google.colab import drive

drive.mount('/content/drive')
````

````python
import os
from IPython.display import Image, display

# Option A: dossier sur Drive
folder = '/content/drive/MyDrive/Data colab/plots-ci'  # adapte

# Option B: dossier dans le repo (après un run)
# folder = 'results/run_csv/plots'
# folder = 'results/run_csv/viz_a_to_h'

assert os.path.isdir(folder), f'Dossier introuvable: {folder}'

png_paths = []
for root, _, files in os.walk(folder):
    for f in files:
        if f.lower().endswith('.png'):
            png_paths.append(os.path.join(root, f))

png_paths = sorted(png_paths)

print(f'Trouvé {len(png_paths)} PNG :')
for p in png_paths:
    print('→', os.path.relpath(p, folder))
    display(Image(filename=p, width=900))
````

````python
import shutil

zip_target = folder  # adapte si besoin
zip_name = 'plots_export'

shutil.make_archive(zip_name, 'zip', zip_target)
print('ZIP créé:', f'{zip_name}.zip')
````

````python
from google.colab import files

files.download('plots_export.zip')
````
