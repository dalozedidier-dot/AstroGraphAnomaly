## Choisir un dossier de résultats

```python
from pathlib import Path
OUT_DIR = Path('results/colab_csv')  # ex: results/colab_gaia
print('Exists:', OUT_DIR.exists())
print('Files:', [p.name for p in OUT_DIR.glob('*')])
```

```python
import pandas as pd
scored = pd.read_csv(OUT_DIR / 'scored.csv')
top = pd.read_csv(OUT_DIR / 'top_anomalies.csv')
print('rows scored:', len(scored))
print('rows top:', len(top))
if 'anomaly_label' in scored.columns:
    print('anomalies (label=-1):', int((scored['anomaly_label'] == -1).sum()))
top.head(10)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
scored['anomaly_score'].hist(bins=60)
plt.title('Anomaly score distribution (higher = more anomalous)')
plt.xlabel('anomaly_score'); plt.ylabel('count')
plt.tight_layout()
plt.show()
```

```python
import json
p = OUT_DIR / 'llm_prompts.jsonl'
if p.exists():
    with p.open('r', encoding='utf-8') as f:
        obj = json.loads(next(f))
    print('source_id:', obj.get('source_id'))
    print(obj.get('prompt')[:1500])
else:
    print('llm_prompts.jsonl not found')
```
